- 코딩 중... 잔고가 소수점 이하 자리까지 내려가는 현상발견...반올림하여 정수로..

3. deposit_history와 withdraw_history 메서드를 통해 기록되고 내역을 출력한다.
4. 계좌의 개수를 출력한다.

```python
import random

class Account:
    account_count = 0   # 전체 계좌 수를 저장
    accounts = []       # 모든 계좌 인스턴스를 저장

    @classmethod
    def check_existing_account(cls, owner_name):  #예금주 이름을 기반으로 이미 존재하는 계좌인지 확인하는 클래스 메서드
        for account in cls.accounts:              #(구현과제에는 없는 것 같은데...기능구현하다가 욕심이 나서 저지름..신설계좌 생성!!!)
            if account.owner == owner_name:
                return True
        return False

    def __init__(self, owner, balance=0):
        if Account.check_existing_account(owner):
            print(f"{owner} 이름으로 이미 계좌가 개설되어 있습니다.")
            return None

        self.bank = "SC은행"       #은행명 초기화
        self.owner = owner        #예금주 초기화
        self.balance = balance    #잔고 초기화 ..기본값 0
        self.deposit_count = 0    #입금 횟수 카운트 초기화, 5회 이상이면 이자 지급해야 함으로...
        self.withdraw_count = 0   #출금 횟수
        self.deposit_log = []     #입금 기록 리스트로 기록관리
        self.withdraw_log = []    #출금 기록 리스트로 기록관리
        self.interest_log = 0     #이자 기록 초기값

        #계좌번호 랜덤 생성 3자리, 2자리, 6자리
        num1 = str(random.randint(100, 999))
        num2 = str(random.randint(10, 99))
        num3 = str(random.randint(100000, 999999))
        self.account_number = num1 + '-' + num2 + '-' + num3
        Account.account_count += 1    # 전체 계좌 수 증가시키고
        Account.accounts.append(self) # 현재 계좌를 계좌 리스트에 추가시킴.

    @classmethod
    def find_account_by_owner(cls, owner_name): #예금주의 이름으로 기존 계좌 검색.
        for acc in cls.accounts:                #신설계좌 예금주와 중복되면 안되므로. 중복이름 신설계좌 불가.
            if acc.owner == owner_name:
                return acc
        return None

    @classmethod
    def get_account_num(cls):     #생성된 계좌의 개수를 출력하는 get_account_num() 메서드를 추가
        return cls.account_count  # 클래스 변수 account_count의 값을 반환

    def get_amount_input(self, prompt_message): ## 사용자로부터 금액을 입력받는 메서드. 잘못된 금액을 입력하면 다시 입력받음
        while True:
            try:
                amount = int(input(prompt_message))
                if amount <= 0:
                    raise ValueError
                return amount
            except ValueError:
                print("----" + "올바른 금액을 입력하세요." + "----")

    ## 입금 메서드
    def deposit(self, amount):
        if amount >= 1:                  #입금은 최소 1원 이상
            self.balance += amount
            self.deposit_log.append(amount)
            self.deposit_count += 1
            if self.deposit_count == 5: # 입금 횟수가 5회, 이자 지급
                interest = round(self.balance * 0.01)  # **** 이자율 계산으로 나중에 잔고가 소수점 이하로 내려가서 반올림 해버림.
                self.balance += interest
                self.interest_log = interest
                self.deposit_count = 0
    # 출금 메서드
    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            self.withdraw_log.append(amount)
            self.withdraw_count += 1
        else:
            print("----잔고가 부족합니다. 출금할 수 없습니다. ----")
    # 계좌 정보 출력 메서드
    def display_info(self):
        print('--'*30)
        print("예금주:", self.owner)
        print("계좌번호:", self.account_number)
        print("잔고:", format(self.balance, ","), "원")
        print("입금기록:", self.deposit_log)
        print(f"이자 기록: 5회 입금되어 잔고의 1%인 {format(self.interest_log, ',')}원이 지급되었습니다.")
        print("입금 횟수 기록:", self.deposit_count)
        print("출금기록:", self.withdraw_log)
        print("출금 횟수 기록:", self.withdraw_count, "\n" + "--"*30)

# 여러 객체 생성
accounts_list = [Account("홍길동", 10000), Account("이순신", 2000000),
                 Account("장보고", 500000), Account("김철수", 10000),
                 Account("이영희", 200000), Account("최돌쇠", 50000)]
# 메인 메뉴 출력
while True:
    print("원하시는 동작을 선택하세요 (🥇계좌신설: 1, 🥈입금: 2, 🥉출금: 3, 🎖️100만원 이상 고객 조회: 4, 🏅계좌 수 조회: 5, ⛩️종료: q)")
    action = input().lower()

    if action == '1':
        while True:
            owner_name = input("🥇계좌를 신설하겠습니다. 이름을 입력하세요: ")
            if owner_name.strip() and not owner_name.isnumeric(): #이름에 숫자넣는 오류 방지
                account = Account(owner_name)
                if hasattr(account, 'account_number'):
                    print(f"계좌번호: {account.account_number}")
                    deposit_amount = account.get_amount_input("입금액을 입력하세요: ")
                    account.deposit(deposit_amount)
                    account.display_info()
                    break
            else:
                print("---- 올바른 예금주 이름을 입력하세요. ----")

    elif action == '2':
        owner_name = input("예금주 이름을 입력하세요: ")
        account = Account.find_account_by_owner(owner_name)
        if account:
            print(f"계좌번호: {account.account_number}")
            deposit_amount = account.get_amount_input("입금액을 입력하세요: ")
            account.deposit(deposit_amount)
            account.display_info()
        else:
            print("---- 해당 예금주의 계좌가 없습니다. ----")

    elif action == '3':
        owner_name = input("예금주 이름을 입력하세요: ")
        account = Account.find_account_by_owner(owner_name)
        if account:
            print(f"계좌번호: {account.account_number}")
            withdraw_amount = account.get_amount_input("출금액을 입력하세요: ")
            account.withdraw(withdraw_amount)
            account.display_info()
        else:
            print("---- 해당 예금주의 계좌가 없습니다.----")

    elif action == '4':
        print("🎖️잔고가 100만원 이상인 고객 정보를 출력합니다.")
        high_balance_accounts = [acc for acc in Account.accounts if acc.balance >= 1000000]
        if high_balance_accounts:
            for acc in high_balance_accounts:
                acc.display_info()
        else:
            print("---- 100만원 이상 고객이 없습니다. ----")
    elif action == '5':
        print('--'*30)
        print(f"🏅현재 생성된 계좌 수: {Account.get_account_num()}개")
        print('--'*30)

    elif action == 'q':
        break
    else:
        print("---- 잘못된 입력입니다. ----")
```
