#🖥 프로젝트1. 사각형 넓이 구하기

```python
class Square:
    def __init__(self):
        while True:  # 무한 루프
            try:     # 예외 처리
                square = int(input('🪧 넓이를 구하고 싶은 도형의 종류를 선택해주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 4.종료\n >>>'))
            except ValueError:    # 예외 처리
                print("😇 숫자를 입력해주세요." + '\n' + '--'*30)
                continue
            #선택시 각 메서드 실행
            if square == 1:
                print(self.rect())

            elif square == 2:
                print(self.par())

            elif square == 3:
                print(self.trape())

            elif square == 4:
                print("😎 프로그램을 종료합니다.")
                break  # 무한 루프를 빠져나갑니다.

            else:
                print('😇 1, 2, 3, 4 중에서 다시 입력해주세요' + '\n' + '--'*30)

    # 평행사변형 메서드
    def rect(self):
        try:
            print('--'*30)
            width, vertical = map(int, input('🧮 사각형의 넓이를 구하기 위해 가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        except ValueError:
            return "😇 올바른 숫자를 입력해주세요." + '\n' + '--'*30

        area = width * vertical
        result = '🥁 직사각형의 넓이는 : ' + str(area) + '\n' + '--'*30
        return result

    # 평행사변형 메서드
    def par(self):
        try:
            print('--'*30)
            lower, height = map(int, input("🧮 평행사변형의 넓이를 구하기 위해 밑변, 높이를 입력해 주세요. 예시: 밑변,높이\n >>>").split(","))
        except ValueError:
            return "😇 올바른 숫자를 입력해주세요." + '\n' + '--'*30

        area = lower * height
        result = "🥁 평행사변형의 넓이는 : " + str(area) + '\n' + '--'*30
        return result

    # 사다리꼴 메서드
    def trape(self):
        try:
            print('--'*30)
            lower, upper, height = map(int, input("🧮 사다리꼴의 넓이를 구하기 위해 밑변, 윗변, 높이를 입력해 주세요. 예시: 밑변,윗변,높이\n >>>").split(","))
        except ValueError:
            return "😇 올바른 숫자를 입력해주세요." + '\n' + '--'*30

        area = (lower + upper) * height / 2
        result = "🥁 사다리꼴의 넓이는 : " + str(area) + '\n' + '--'*30
        return result

# 객체 생성 & 메뉴 선택
a = Square()
```

# 🖥 프로젝트2. kiosk 만들기


---


```python
# kiosk 만들기

# 메뉴와 가격 리스트를 다차원 리스트로 변경하지 않았습니다.실력도 부족하고 필요성을 못 느꼈습니다. 혹 다층메뉴 선택 기회시 도전해 보려고 합니다.
# 메뉴 출력 enumerate() 함수를 활용하였습니다.
# datetime 모듈로 주문표에 주문 일시 추가하였습니다.
# 데코레이터로 주문표 양식 출력과 메뉴 양식 출력을 더했습니다.
# 예외처리 (메뉴/온도 선택시 빈값 엔터, 선택번호 벗어난 값 입력시...)를 하였습니다.

menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
price = [2000, 3000, 3000, 2500, 2500, 3000]

from datetime import datetime

class Kiosk:
    def __init__(self, menu, price):
        self.menu = menu
        self.price = price

    def menu_print_decorator(func):           #메뉴 전체를 박스로 감싸는 데코레이터 함수
        def wrapper(self, *args, **kwargs):
            print('⟝' + '-' * 30 + '⟞')      # 박스 상단 부분
            for i in range(2):
                print('|' + ' ' * 31 + '|')   # 박스의 빈 공간을 출력

            func(self, *args, **kwargs)       # 원래 함수를 호출

            for i in range(2):
                print('|' + ' ' * 31 + '|')   # 박스의 빈 공간을 출력
            print('⟝' + '-' * 30 + '⟞')
        return wrapper                        #내부 래퍼 함수를 반환

    @menu_print_decorator                     # 메뉴 데코레이터
    def menu_print(self):                     # 메뉴 출력 메서드
        for i, item in enumerate(self.menu):  # 메뉴 출력 enumerate() 함수 활용
            print(f"| {i + 1}. {item} : {self.price[i]} 원".ljust(30))

    def menu_select(self):                    #주문 메서드
        self.order_menu = []                  #주문 리스트
        self.order_price = []                 #가격 리스트
        self.price_sum = 0

        while True:
            try:                              #예외처리 (메뉴선택시)
                n = int(input("음료 번호를 입력하세요 : ")) #음료 번호 입력
                if 1 <= n and n <= len(self.menu):
                    self.order_price.append(self.price[n-1])
                    self.price_sum += self.price[n-1]
                    break
                else:                         #메뉴판에 없는 번호일 때
                    print("--없는 메뉴입니다. 다시 선택해 주세요--")
            except ValueError:
                print("---- 아무것도 입력하지 않았습니다 ----")

        while True:                           #음료 온도 물어보기
            try:                              #예외처리 (HOT or ICE 선택)
                t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                if t == 1:
                    temp = 'HOT'
                    break
                elif t == 2:
                    temp = 'ICE'
                    break
                else:
                    print("-----1과 2 중 하나만 선택하세요-----\n")
            except ValueError:
                print("---- 아무것도 입력하지 않았습니다 ----")

        self.order_menu.append(temp + ' ' + self.menu[n-1]) # 주문 리스트에 추가
        print("-"*30)                         #구분선
        print(f'주문음료 : {self.order_menu[-1]} : {self.price_sum} 원\n')

        while True:
            try:
                n = int(input("추가 주문은 음료 번호를, 지불은 0을 누르세요 : "))
                if n > 0 and n < len(self.menu) + 1:
                    self.order_price.append(self.price[n-1])  # 가격 리스트에 추가합니다.
                    self.price_sum += self.price[n-1]         # 합계 금액 업데이트

                    while True:
                        try:
                            t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                            if t == 1:
                                temp = 'HOT'
                                break
                            elif t == 2:
                                temp = 'ICE'
                                break
                            else:
                                print("-----1과 2 중 하나만 선택하세요-----\n")
                        except ValueError:
                            print("---- 아무것도 입력하지 않았습니다 ----")

                    self.order_menu.append(temp + ' ' + self.menu[n-1])
                    print('-'*30, '\n','추가 주문 음료 : ', self.order_menu[-1], ':', self.price[n-1], '원\n', '합계 : ', self.price_sum, '원', '\n', '-'*30, '\n' )
                elif n == 0:                # 지불을 입력하면
                    print("-"*30, "\n","주문이 완료되었습니다.\n", "-"*30,)
                    formatted_orders = [f"{menu} : {price}원" for menu, price in zip(self.order_menu, self.order_price)]  # 확인을 위한 리스트를 출력합니다. HOT americano : 2000원 형식으로 출력하기 위해 ..기존 리스트를 zip함수를 이용해서 재배열
                    print(", ".join(formatted_orders))
                    break
                else:                       # 없는 번호를 입력할 때
                    print("--없는 메뉴입니다. 다시 선택해 주세요--")
            except ValueError:
                print("---- 아무것도 입력하지 않았습니다 ----")

    def pay(self):                          # 지불
        print("지불할 금액:", self.price_sum, "원\n", "-"*30)    # 지불할 금액 출력
        payment = 0                         # 초기 선택 값
        while payment not in [1, 2, 'cash', 'card']:
            payment = input("지불 수단을 선택하세요 (현금: 'cash' 또는 1, 카드: 'card' 또는 2) : ")
            if payment in ['cash', '1']:
                print('직원을 호출하겠습니다.')
                break
            elif payment in ['card', '2']:
                print('IC칩 방향에 맞게 카드를 꽂아주세요.')
                break
            else:                           #입력값없이 엔터를 치거나 다른 1,2 이외 입력값을 넣었을 경우. 예외처리.
                print("----- 다시 결제를 시도해 주세요. 1과 2 중 하나만 선택하세요 -----")

    @menu_print_decorator                   #데코레이터. 주문출력 전체를 박스 처리.
    def table(self):
        for i in range(2):
            print('|' + ' ' * 31 + '|')
        now = datetime.now()
        print('|', f"주문일시: {now.strftime('%Y-%m-%d %H:%M:%S')}") # datetime 모듈로 주문표에 주문 일시 추가하기
        for i in range(2):
            print('|' + ' ' * 31 + '|')
        for i in range(len(self.order_menu)):
            print('|', self.order_menu[i].ljust(15), ':', str(self.order_price[i]).rjust(5), '원')
        print('|', '합계 금액'.ljust(13), ':', str(sum(self.order_price)).rjust(5), '원')
        for i in range(5):
            print('|' + ' ' * 31 + '|')


a = Kiosk(menu, price) #객체 생성
a.menu_print()         #메뉴 출력
a.menu_select()        #주문
a.pay()                #지불
a.table()              #주문표 출력

```
